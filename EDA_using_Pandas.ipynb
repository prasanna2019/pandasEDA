```python
import pandas as pd
import plotly.express as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasanna2019/pandasEDA/main/TFL%20Bus%20Safety.csv', parse_dates=['Date Of Incident'])
df.head(5)
```

# Let us understand the data we have.

```python
df.dtypes
```

```python
df.info()
```

**Number of incidents happened in each year**

```python
data= df.groupby(['Year']).size().reset_index(name='Incidents')
data.head()
```

# Incidents occurring each year
Let's see a line graph of incidents occured per year.

```python
graph= plt.line(data,x="Year", y= "Incidents", title="Year vs Incidents")
graph.update_layout(xaxis=dict(tickmode='linear'))
graph.show()

```

## As seen from the data above, most incidents occurred in the year 2017


Similarly we can find which operator, borough, victim sex and age has most number of incidents.





```python
type_victim= df.groupby('Victim Category').size()
type_victim.idxmax()
```

```python
type_victim.max()
```

**Maximum accidents occur to Passenger victim type, which are 18828**

```python
df['day']= df['Date Of Incident'].dt.day_name()
incindents_by_day= df.groupby('day').size()
incindents_by_day
```

```python
incindents_by_day.idxmax(), incindents_by_day.max()
```

```python
incindents_by_day.idxmin(), incindents_by_day.min()
```

**As seen, most incidents occur on Monday and least occur on Wednesday**

```python
ages=df.groupby('Victims Age').size().reset_index(name='Count')
ages
```

```python
fig = plt.pie(ages, names='Victims Age', values='Count', title='Category1 Distribution')
fig.show()
```

**Adults and unkown are the two age groups most affected in the incidents and Youth are the least affected.**
